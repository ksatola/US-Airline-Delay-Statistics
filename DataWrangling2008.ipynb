# Assessing & Cleaning

```python
#df.DepTime
#df[df.DepTime.isna()]
# When a flight is cancelled, DepTime and ArrTime are NaN
# But not always
df[(~df.DepTime.notnull() | ~df.ArrTime.notnull()) & (df.Cancelled == 0)].count()
```

```python
# Remove rows where there is no DepTime or ArrTime but the flight was not cancelled
mask = (~df.DepTime.notnull() | ~df.ArrTime.notnull()) & (df.Cancelled == 0)
df = df[~mask]
```

```python
df.info(null_counts=True)
```

```python
# Remove rows where there is no DepTime or ArrTime
mask = ((~df.DepTime.notnull() & df.ArrTime.notnull()) | (df.DepTime.notnull() & ~df.ArrTime.notnull())) & (df.Cancelled == 0)
df[mask]
```

```python
df.drop(['CRSDepTime', 
         'CRSArrTime',
         'TailNum',
         'FlightNum', 
         'TaxiIn',
         'TaxiOut',
         'ActualElapsedTime',
         'CRSElapsedTime',
         'AirTime'], axis=1, inplace=True)
```
